### Mas Columnas 

```python
genero_set = set()
for lst in Datos_Movie['Genre'].str.split(','):
    genero_set.update(lst)
```

```python
for g in genero_set:
    Datos_Movie[g] = np.nan
    Datos_Movie.loc[Datos_Movie['Genre'].str.contains(g),g] = g
```

```python
Datos_Movie
```

#### Crear mas columnas propias

```python
def rating_good_bad(x):
    if x >= 8.0:
        return "GOOD"
    else:
        return "BAD"
```

```python
# Crea una nueva columna
Datos_Movie['Rating Category'] = Datos_Movie['Rating'].apply(rating_good_bad)
```

```python
Datos_Movie
```

```python
def rentabilidad(ret):
    if ret >= 500:
        return "Rentable"
    else:
        return "No rentable"
```

```python
Datos_Movie["Rentabilidad"] = Datos_Movie['Revenue (Millions)'].apply(rentabilidad)
Datos_Movie
```
